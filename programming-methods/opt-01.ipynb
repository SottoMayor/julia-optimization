Seja $x_0 = [-0.5, 1.0]^T$ e a direção $p = [-1, -1.5]^T$. O algoritmo backtracking line Search para $\underline{\alpha}=1, \rho=0.5, c_1 = 1E-4$, resulta em $ \alpha = 0.25$
Teste a sua implementação do algoritmo e confira se chegou no mesmo resultado. Crie uma função chamada backtracking_line_search que recebe como parâmetros: 

* f: calcula a função num dado ponto x
* df: calcula a derivada da função num dado ponto x
* x: ponto de interesse
* alpha: tamanho do passo inicial
* rho: fator de escalonamento do backtracking line search
* c1: constante utilizada na condição de decréscimo suficiente

```julia
function backtracking_line_search(f, df, x, p, alpha, rho = 0.5, c1 = 1e-4)
    # valor da função e da derivada no ponto
    fx = f(x)
    dfx = df(x)
    
    # backtracking line search
    while f(x + alpha * p) > fx + c1 * alpha * dot(dfx, p)
        alpha *= rho
    end
    
    return alpha
end

alpha_ = backtracking_line_search(f, df, [-0.5, 1.0], [-1, -1.5], 1.0)
print("Alpha: ", alpha_)
```

Utilizando como ponto inicial $x_0 = [-5., 0]^T$ e direção $p_0 = [0.9958, -0.0917]^T$, encontre um tamanho de passo $\alpha$ que satisfaça as condições fortes de Wolfe. Use $c_1 = 1E-4, c_2 = 0.9$.

Plote a função $\phi (\alpha)$ no intervalo $[0, 11]$. No mesmo gráfico, plote a função $l(\alpha) = f(x_k) + c_1 \alpha \nabla f_k^T p_k$

```julia
x0 = [-5, 0]
p = [0.9958, -0.0917]
c1 = 1E-4 
c2 = 0.9
alpha_init = 1.0

# Condições fortes de Wolfe
function strong_wolfe_line_search(f, df, x, p, c1=1E-4, c2=0.9, alpha_init=1.0)
    alpha = alpha_init
    alpha_prev = 0.0
    max_iter = 100
    iter = 0
    
    while iter < max_iter
        iter += 1
        
        # decréscimo suficiente
        if f(x + alpha * p) > f(x) + c1 * alpha * dot(df(x), p)
            alpha_prev = alpha
            alpha = 0.5 * (alpha + alpha_prev)
            continue
        end
        
        # condição de curvatura
        if abs(dot(df(x + alpha * p), p)) > c2 * abs(dot(df(x), p))
            alpha_prev = alpha
            alpha = 0.5 * (alpha + alpha_prev)
            continue
        end
        
        # Se tiver tudo ok, retorna alpha
        return alpha
    end
    
    return alpha  # se não convergir, retorna o alpha inicial ou encontrado.
end

alpha = strong_wolfe_line_search(f, df, x0, p)

println("Tamanho do passo usando as condições fortes de Wolfe: ", alpha)
```

```julia
phi(a) = f(x0 + a * p)
l(a) = f(x0) + c1 * a * dot(df(x0), p)

X = range(0,11,100)

Z = [phi(a) for a in X]
Y = [l(a) for a in X]

pyplt.figure()
pyplt.plot(X, Z)
pyplt.plot(X,Y, "k-.")
pyplt.xlabel("\$ α \$")
pyplt.grid(true)
```

## 4 - Exemplo 

Seja a função

$$ f(x) = x_1^2 + x_1x_2 + x_2^2$$

Utilizando como ponto inicial $x_0 = [1, 2]^T$ e direção $p_0 = [-1., -1.]^T$, realize a busca linear inexata que satisfaça as condições de Wolfe.

Use como parâmetros: $\underline{\alpha} = 10, \rho = 0.5, c_1 = 1E-4, c_2 = 0.9$. 

Neste exemplo, as condições de Wolfe são atendidas na 3a iteração, com $\alpha = 2.5, x_1 = [-1.5, -0.5]^T$. Use o pacote `ForwardDiff`para calcular as derivadas.

```julia
import PyPlot as pyplt
using LinearAlgebra
using ForwardDiff

# função objetivo
function f(x)
    return x[1]^2 + x[1]*x[2] + x[2]^2
end

# gradiente usando ForwardDiff
function df(x)
    return ForwardDiff.gradient(f, x)
end

# verificação das condições de Wolfe
function verifica_condicoes_wofle(f, df, x0, p, alpha, c1, c2)
    phi_alpha = f(x0 + alpha * p)
    phi_0 = f(x0)
    grad_phi_0 = df(x0)
    grad_phi_alpha = df(x0 + alpha * p)
    
    decrescimento_suficiente = phi_alpha <= phi_0 + c1 * alpha * dot(grad_phi_0, p)
    cond_curvatura = dot(grad_phi_alpha, p) >= c2 * dot(grad_phi_0, p)
    
    return decrescimento_suficiente, cond_curvatura
end

# busca linear
function busca_linear(f, df, x0, p, alpha0, rho, c1, c2)
    alpha = alpha0
    iter = 0
    while iter < 100
        decrescimento_suficiente, cond_curvatura = verifica_condicoes_wofle(f, df, x0, p, alpha, c1, c2)
        if decrescimento_suficiente && cond_curvatura
            return alpha
        end
        alpha *= rho
        iter += 1
    end
    error("Busca linear não convergiu")
end

x0 = [1.0, 2.0]
p0 = [-1.0, -1.0]
alpha0 = 10.0
rho = 0.5
c1 = 1E-4
c2 = 0.9

alpha_opt = busca_linear(f, df, x0, p0, alpha0, rho, c1, c2)

x1 = x0 + alpha_opt * p0


println("Alpha ótimo: ", alpha_opt)
println("Próximo ponto x1: ", x1)

# Gráficos

phi(a) = f(x0 + a * p0)
l(a) = f(x0) + c1 * a * dot(df(x0), p0)

X = range(0, 10, length=100)
Z = [phi(a) for a in X]
Y = [l(a) for a in X]

pyplt.figure()
pyplt.plot(X, Z, label="phi(alpha)")
pyplt.plot(X, Y, "k-.", label="l(alpha)")
pyplt.xlabel("alpha")
pyplt.grid(true)
pyplt.legend()
pyplt.show()
```

```julia
using LinearAlgebra

function f(x)
    return 5*x[1]^2 + x[2]^2 + 4*x[1]*x[2] - 14*x[1] - 6*x[2] + 20
end

function df(x)
    return ForwardDiff.gradient(f, x)
end

function busca_linear_exata(f, df, x, p)
    phi(alpha) = f(x + alpha * p)
    alpha_opt = optimize(phi, 0, 1).minimizer
    return alpha_opt
end

function gradiente_busca_linear(f, df, x0, tol, max_iter)
    x = x0
    iter = 0
    while norm(df(x)) > tol && iter < max_iter
        grad = df(x)
        p = -grad
        alpha = busca_linear_exata(f, df, x, p)
        x = x + alpha * p
        iter += 1
        println("Iteração $iter: x = $x, f(x) = $(f(x))")
    end
    return x, f(x), iter
end

x0 = [0.0, 0.0]
tol = 1E-6
max_iter = 100

# OBS: para a tolerância 1E-6 o meu deu mais de 24, deu 74 pra ser exato.
gradiente_busca_linear(f, df, x0, tol, max_iter)
```

Agora refaça considerando método do gradiente e busca linear inexata com backtracking line search.

```julia
using LinearAlgebra
using ForwardDiff

function f(x)
    return 5*x[1]^2 + x[2]^2 + 4*x[1]*x[2] - 14*x[1] - 6*x[2] + 20
end

function df(x)
    return ForwardDiff.gradient(f, x)
end

# condição de decrescimento suficiente
function condicao_decrescimento_suficiente(f, df, x, p, alpha, c1)
    phi_alpha = f(x + alpha * p)
    phi_0 = f(x)
    grad_phi_0 = df(x)
    return phi_alpha <= phi_0 + c1 * alpha * dot(grad_phi_0, p)
end

#  busca linear inexata com backtracking line search
function backtracking_line_search(f, df, x, p, alpha0, rho, c1)
    alpha = alpha0
    while !condicao_decrescimento_suficiente(f, df, x, p, alpha, c1)
        alpha *= rho
    end
    return alpha
end

# método do gradiente com busca linear inexata
function gradiente_busca_linear_inexata(f, df, x0, tol, max_iter, alpha0, rho, c1)
    x = x0
    iter = 0
    while norm(df(x)) > tol && iter < max_iter
        grad = df(x)
        p = -grad
        alpha = backtracking_line_search(f, df, x, p, alpha0, rho, c1)
        x = x + alpha * p
        iter += 1
        println("Iteração $iter: x = $x, f(x) = $(f(x))")
    end
    return x, f(x), iter
end

x0 = [0.0, 0.0]
tol = 1E-6
max_iter = 1000
alpha0 = 1.0
rho = 0.5
c1 = 1E-4

# A diferença foi absurda. Esse método precisou de 116 iterações a mais para 
# chegar no mesmo resultado que o método anterior.
gradiente_busca_linear_inexata(f, df, x0, tol, max_iter, alpha0, rho, c1)
```

```julia
import PyPlot as pyplt
using LinearAlgebra
using ForwardDiff

function f(x)
    return sum([(2*x[i+1] - x[i]^2)^2 + (1 - x[i])^2 for i in 1:2])
end

function df(x)
    return ForwardDiff.gradient(f, x)
end

function condicao_decrescimento_suficiente(f, df, x, p, alpha, c1)
    phi_alpha = f(x + alpha * p)
    phi_0 = f(x)
    grad_phi_0 = df(x)
    return phi_alpha <= phi_0 + c1 * alpha * dot(grad_phi_0, p)
end

function condicao_curvatura(df, x, p, alpha, c2)
    grad_phi_alpha = df(x + alpha * p)
    grad_phi_0 = df(x)
    return abs(dot(grad_phi_alpha, p)) <= c2 * abs(dot(grad_phi_0, p))
end

function backtracking_line_search(f, df, x, p, alpha0, rho, c1, c2)
    alpha = alpha0
    while !condicao_decrescimento_suficiente(f, df, x, p, alpha, c1) || !condicao_curvatura(df, x, p, alpha, c2)
        alpha *= rho
    end
    return alpha
end

function gradiente_busca_linear_inexata(f, df, x0, tol, max_iter, alpha0, rho, c1, c2)
    x = x0
    iter = 0
    hist = []
    while norm(df(x), Inf) > tol && iter < max_iter
        grad = df(x)
        p = -grad
        alpha = backtracking_line_search(f, df, x, p, alpha0, rho, c1, c2)
        x = x + alpha * p
        iter += 1
        push!(hist, (iter, copy(x), f(x), norm(grad, Inf), alpha))
    end
    return x, f(x), iter, hist
end

x0 = [-1.2, 1.0, 0.0]
tol = 1E-4
max_iter = 1000
alpha0 = 1.0
rho = 0.5
c1 = 1E-4
c2 = 0.9

x_opt, f_opt, n_iter, hist = gradiente_busca_linear_inexata(f, df, x0, tol, max_iter, alpha0, rho, c1, c2)

println("Em $n_iter Iterações:" )
println("x* = $x_opt")
println("f(x*) = $f_opt\n\n")

println("###################################################################################\n\n")

function phi(alpha)
    return f(x0 + alpha * (-df(x0)))
end

alphas = range(0, 2, length=100)
phis = [phi(alpha) for alpha in alphas]

figure()
plot(alphas, phis, label="φ(α)")
xlabel("α")
ylabel("φ(α)")
title("Função φ(α) na primeira iteração")
grid(true)
legend()
show()

# função auxiliar para exibir dados
function imprimir_tabela(hist)
    println("Iteração | x | f(x) | ||∇f(x)|| | α")
    println("---------------------------------------------------------")
    for (iter, x, fx, norm_grad, alpha) in hist
        println("$iter | $x | $fx | $norm_grad | $alpha")
    end
end

imprimir_tabela(hist)

println("\n\n###################################################################################\n\n")

iters = [h[1] for h in hist]
f_values = [h[3] for h in hist]
norm_grad_values = [norm(df(h[2])) for h in hist]
alpha_values = [h[5] for h in hist]

figure()
subplot(3,1,1)
plot(iters, f_values, marker="x")
xlabel("Iteração")
ylabel("f(x)")
title("f(x) ao longo das iterações")
grid(true)

subplot(3,1,2)
plot(iters, norm_grad_values, marker=".")
xlabel("Iteração")
ylabel("||∇f(x)||")
title("||∇f(x)|| ao longo das iterações")
grid(true)

subplot(3,1,3)
plot(iters, alpha_values, marker="o")
xlabel("Iteração")
ylabel("α")
title("α ao longo das iterações")
grid(true)

# Ajusta o espaçamento entre os subplots
subplots_adjust(hspace=2)

show()

```

Cujo mínimo global é $x^* = [1., 1.,]^T$. Implemente o método de Newton com backtracking line search. Utilize como tamanho de passo inicial $\alpha_0 = 1$ e imprima o tamanho do passo utilizado em cada iteração. Tente inicialmente com o ponto inicial $x_0 = [1.2, 1.2]^T$ e depois tente com um ponto mais difícil $x_0 = [-1.2, 1.0]^T$.

```julia
import PyPlot as pyplt
using LinearAlgebra
using ForwardDiff

function rosenbrock(x)
    return 100*(x[2] - x[1]^2)^2 + (1 - x[1])^2
end

function grad_rosenbrock(x)
    return ForwardDiff.gradient(rosenbrock, x)
end

function hess_rosenbrock(x)
    return ForwardDiff.hessian(rosenbrock, x)
end

function condicao_decrescimento_suficiente(f, df, x, p, alpha, c1)
    phi_alpha = f(x + alpha * p)
    phi_0 = f(x)
    grad_phi_0 = df(x)
    return phi_alpha <= phi_0 + c1 * alpha * dot(grad_phi_0, p)
end

function condicao_curvatura(df, x, p, alpha, c2)
    grad_phi_alpha = df(x + alpha * p)
    grad_phi_0 = df(x)
    return abs(dot(grad_phi_alpha, p)) <= c2 * abs(dot(grad_phi_0, p))
end

function backtracking_line_search(f, df, x, p, alpha0, rho, c1, c2)
    alpha = alpha0
    while !condicao_decrescimento_suficiente(f, df, x, p, alpha, c1) || !condicao_curvatura(df, x, p, alpha, c2)
        alpha *= rho
    end
    return alpha
end

function newton_backtracking(f, df, hess, x0, tol, max_iter, alpha0, rho, c1, c2)
    x = x0
    iter = 0
    hist = []
    while norm(df(x), Inf) > tol && iter < max_iter
        grad = df(x)
        H = hess(x)
        p = -H \ grad
        alpha = backtracking_line_search(f, df, x, p, alpha0, rho, c1, c2)
        x = x + alpha * p
        iter += 1
        push!(hist, (iter, copy(x), f(x), norm(grad, Inf), alpha))
    end
    return x, f(x), iter, hist
end

x0_a = [1.2, 1.2]
x0_b = [-1.2, 1.0]
tol = 1E-6
max_iter = 1000
alpha0 = 1.0
rho = 0.5
c1 = 1E-4
c2 = 0.9

println("\nPonto inicial: [1.2, 1.2]\n")
x_opt_a, f_opt_a, n_iter_a = newton_backtracking(rosenbrock, grad_rosenbrock, hess_rosenbrock, x0_a, tol, max_iter, alpha0, rho, c1, c2)

println("Em $n_iter_a Iterações:" )
println("x* = $x_opt_a")
println("f(x*) = $f_opt_a")

println("###################################################################################")

println("\nPonto inicial: [-1.2, 1.0]\n")
x_opt_b, f_opt_b, n_iter_b = newton_backtracking(rosenbrock, grad_rosenbrock, hess_rosenbrock, x0_b, tol, max_iter, alpha0, rho, c1, c2)

println("Em $n_iter_b Iterações:" )
println("x* = $x_opt_b")
println("f(x*) = $f_opt_b\n\n")

```
