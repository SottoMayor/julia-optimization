```julia
# Trabalho Final - Item 2

using JuMP
using HiGHS

# valores de yield de cada alimento, considerando o cenário médio
yield_trigo = 2.5
yield_milho = 3.0
yield_beterraba = 10.0

# geração dos 10 cenários espaçados, ao intervalo de -20% e 20% dos valores de cada yield.
fatores = range(0.8, 1.2, length=10)
cenarios = [(yield_trigo * f, yield_milho * f, yield_beterraba * f) for f in fatores]

model = Model(HiGHS.Optimizer)
set_silent(model)

# variáveis, onde x são as de 1° estágio
@variable(model, x1 >= 0)
@variable(model, x2 >= 0)
@variable(model, x3 >= 0)
@variable(model, y[1:10, 1:2] >= 0)
@variable(model, w[1:10, 1:4] >= 0)

# função objetivo
@objective(model, Min,
    150*x1 + 230*x2 + 260*x3 + 
    sum(1/10 * (238*y[s,1] - 170*w[s,1] + 210*y[s,2] - 150*w[s,2] - 36*w[s,3] - 10*w[s,4]) for s in 1:10)
)

# Restrições para todos os 10 cenários
for s in 1:10
    @constraint(model, x1 + x2 + x3 <= 500)
    @constraint(model, cenarios[s][1]*x1 + y[s,1] - w[s,1] >= 200)
    @constraint(model, cenarios[s][2]*x2 + y[s,2] - w[s,2] >= 240)
    @constraint(model, w[s,3] + w[s,4] <= cenarios[s][3]*x3)
    @constraint(model, w[s,3] <= 6000)
end

optimize!(model)

optimal_value = -1 * round(objective_value(model), digits=2)
println("Valor ótimo da para os 10 cenários: R\$ $optimal_value\n")

println("""
Resultado: 
 
    Valor ótimo para os 10 cenários: R\$ 62.374,29
    Valor ótimo para os 3 cenários: R\$ 107427.0
    Diferença em relação ao 2 cenários: R\$ -45.052,71
    
    Observa-se uma diferença considerável entre os modelos de 3 e 10 cenários; a discrepância é de R\$ 45052.71, isto é,
    uma diferença de aproximadamente 72,3%. Essa diferença indica que há impactos ao considerar mais cenários na modelagem.
    
""")
```

```julia
# Trabalho Final - Item 3

using JuMP
using HiGHS

# valores de yield de cada alimento, considerando o cenário médio
yield_trigo = 2.5
yield_milho = 3.0
yield_beterraba = 10.0

# aqui é onde se controla a criação dos cenários, novamente igualmente espaços em relação ao intervalo de 
#-20% e 20% dos valores de cada yield.
for num_scenarios in 246:266
    fatores = range(0.8, 1.2, length=num_scenarios)
    cenarios = [(yield_trigo * f, yield_milho * f, yield_beterraba * f) for f in fatores]

    model = Model(HiGHS.Optimizer)
    set_silent(model)
    
    # variáveis
    @variable(model, x1 >= 0)
    @variable(model, x2 >= 0)
    @variable(model, x3 >= 0)
    @variable(model, y[1:num_scenarios, 1:2] >= 0)
    @variable(model, w[1:num_scenarios, 1:4] >= 0)

    @objective(model, Min,
        150*x1 + 230*x2 + 260*x3 + 
        sum(1/num_scenarios * (238*y[s,1] - 170*w[s,1] + 210*y[s,2] - 150*w[s,2] - 36*w[s,3] - 10*w[s,4]) for s in 1:num_scenarios)
    )

    # Restrições para todos os cenários gerados, observe que no "for num_scenarios" estou gerando apenas de 20 em 20 para
    # não poluir os logs e exigir menos computacionalmente.
    for s in 1:num_scenarios
        @constraint(model, x1 + x2 + x3 <= 500)
        @constraint(model, cenarios[s][1]*x1 + y[s,1] - w[s,1] >= 200)
        @constraint(model, cenarios[s][2]*x2 + y[s,2] - w[s,2] >= 240)
        @constraint(model, w[s,3] + w[s,4] <= cenarios[s][3]*x3)
        @constraint(model, w[s,3] <= 6000)
    end

    optimize!(model)

    optimal_value = -1 * round(objective_value(model), digits=2)
    println("Cenário: $num_scenarios, Valor ótimo: $optimal_value")
end


println("""
         \n
Resultado: 

    Percebeu-se que a partir do cenário 246 a parte inteira do valor da função objetivo estabilizou-se em R\$ 62441, 
    a partir daí as casas decimais sofrem pequenas alterações.
""")
```
